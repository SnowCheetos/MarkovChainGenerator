```python
import numpy as np
import pandas as pd
from src import agents, graph
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("examples/heart_rate.csv")
```

```python
# Visualizing heart rate data
plt.plot(data)
plt.xlabel("Time elapsed")
plt.ylabel("Beats Per Minute")
plt.show()
```

```python
bpm = np.array(data)
n = int(0.8 * len(bpm))
```

```python
#splits the data into training and testing
train, test = bpm[:n], bpm[n:]
```

```python
# Calculates transition space, or an unique set of the training set
t_space = graph.calc_transition_space(train)
```

```python
# Constructs the Markov transition matrix from training set
M = graph.make_graph(train, t_space)
```

```python
# Visualizing the Markov transition matrix
plt.imshow(M)
plt.show()
```

```python
# Perform random walks given first element of testing set
predicted = agents.random_walks(M, test[0], len(test), t_space)
```

```python
# Visualizing predicted results compared to actual testing set
plt.plot(predicted)
plt.plot(test)
plt.legend(["Predicted", "True"])
plt.xlabel("Time elapsed")
plt.ylabel("Beats Per Minute")
plt.show()
```
